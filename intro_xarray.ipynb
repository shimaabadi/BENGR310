Xarray is a powerful Python library designed for working with labeled multidimensional arrays, specifically for working with netCDF files. It provides tools for working with multi-dimensional arrays, as well as time-series data, and allows for easy manipulation and analysis of large datasets.

# Creating a DataArray

```python
import numpy as np
import xarray as xr

# create a DataArray with random values and labeled dimensions
data = np.random.rand(4, 3)
da = xr.DataArray(data, dims=('time', 'space'), coords={'time': [1, 2, 3, 4], 'space': ['x', 'y', 'z']})

print(da)
```

# Indexing and Slicing

```python
# create a DataArray with labeled dimensions
data = np.random.rand(4, 3, 2)
da = xr.DataArray(data, dims=('time', 'space', 'depth'), coords={'time': [1, 2, 3, 4], 'space': ['x', 'y', 'z'], 'depth': [0, 1]})

# select a single value using label-based indexing
print(da.loc[{'time': 2, 'space': 'y', 'depth': 0}])
```

```python
# select a slice using position-based indexing
print(da.isel(time=slice(1, 3), space=0))
```

```python
# select values based on a condition
print(da.where(da > 0.5, other=0))
```

```python
da.where(da > 0.5).plot()
```

```python
da.loc[{'time': 2}].plot()
```

# Basic Operations

```python
# create two DataArrays with random values
data1 = np.random.rand(4, 3)
da1 = xr.DataArray(data1, dims=('time', 'space'), coords={'time': [1, 2, 3, 4], 'space': ['x', 'y', 'z']})

data2 = np.random.rand(4, 3)
da2 = xr.DataArray(data2, dims=('time', 'space'), coords={'time': [1, 2, 3, 4], 'space': ['x', 'y', 'z']})
```

```python
# perform basic mathematical operations
print(da1 + da2)
print(da1.mean())
```

```python
# apply aggregation functions
print(da1.max(dim='time'))
```
