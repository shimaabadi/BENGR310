```python
from PIL import Image

# Load the image
image = Image.open('cat.jpg')

# Display the image
image.show()
```

```python
# Convert the image to grayscale
gray_image = image.convert('L')

# Display the grayscale image
gray_image.show()

#gray_image.save("gray_image.jpg")
```

```python
# rotating
rotated_image = image.rotate(60)
rotated_image
```

```python
from PIL import ImageFilter

blurred_image = image.filter(ImageFilter.BLUR)
blurred_image
```

```python
sharpened_image_2 = blurred_image.filter(ImageFilter.UnsharpMask(radius=2,  percent=130, threshold=1))

sharpened_image_2
```

```python
sharpened_image = image.filter(ImageFilter.SHARPEN)

sharpened_image
```

```python
blurred_image_gaussian = image.filter(ImageFilter.GaussianBlur(10))
blurred_image_gaussian
```

```python
edge_image = image.filter(ImageFilter.FIND_EDGES)
edge_image
```

```python
gray_edge_image = edge_image.convert('L')
gray_edge_image
```

```python
edge_gray_image = gray_image.filter(ImageFilter.FIND_EDGES)
edge_gray_image
```

```python
contour_image = image.filter(ImageFilter.CONTOUR)
contour_image
```
