In this example, you will work with the Iris dataset, apply Principal Component Analysis (PCA) to reduce the dimensionality of the dataset, reconstruct the original features from the principal components, and calculate the Mean Squared Error (MSE) between the original features and the reconstructed features.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error

# Load the Iris dataset
iris_data = load_iris().data

# Apply PCA
pca = PCA(n_components=2)
iris_pca = pca.fit_transform(iris_data)

# Reconstruct the original features
iris_reconstructed = pca.inverse_transform(iris_pca)

# Calculate the Mean Squared Error (MSE)
mse = mean_squared_error(iris_data, iris_reconstructed)

print("Mean Squared Error (MSE):", mse)

# Plot the original features, reconstructed features, principal components, and component coefficients
plt.figure(figsize=(12, 8))

# Plot the original features
plt.subplot(2, 2, 1)
plt.plot(iris_data[:, 0], label='Feature 1')
plt.plot(iris_data[:, 1], label='Feature 2')
plt.plot(iris_data[:, 2], label='Feature 3')
plt.plot(iris_data[:, 3], label='Feature 4')
plt.xlabel('Samples')
plt.ylabel('Feature Value')
plt.title('Original Features')
plt.legend()

# Plot the reconstructed features
plt.subplot(2, 2, 2)
plt.plot(iris_reconstructed[:, 0], label='Reconstructed Feature 1')
plt.plot(iris_reconstructed[:, 1], label='Reconstructed Feature 2')
plt.plot(iris_reconstructed[:, 2], label='Reconstructed Feature 3')
plt.plot(iris_reconstructed[:, 3], label='Reconstructed Feature 4')
plt.xlabel('Samples')
plt.ylabel('Feature Value')
plt.title('Reconstructed Features')
plt.legend()

# Plot the principal components
plt.subplot(2, 2, 3)
plt.scatter(iris_pca[:, 0], iris_pca[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Principal Components')

# Plot the component coefficients
plt.subplot(2, 2, 4)
plt.plot(range(len(pca.components_[0])), pca.components_[0], 'bo-', label='Component 1')
plt.plot(range(len(pca.components_[1])), pca.components_[1], 'ro-', label='Component 2')
plt.xlabel('Feature Index')
plt.ylabel('Coefficient Value')
plt.title('Component Coefficients')
plt.xticks(range(len(pca.components_[0])), ['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4'])
plt.legend()
```
