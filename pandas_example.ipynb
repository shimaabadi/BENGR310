```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
url = 'https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.csv'
df = pd.read_csv(url)
```

```python
print(df.head())
```

```python
df = df[df['mag'] >= 4]
```

```python
# Create a geometry column from the latitude and longitude columns
geometry = gpd.points_from_xy(df['longitude'], df['latitude'])

# Create a GeoDataFrame from the earthquake data and the geometry column
gdf = gpd.GeoDataFrame(df, geometry=geometry)

# Set the coordinate reference system to WGS84
#gdf.crs = 'EPSG:4326'
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
ax = world.plot(figsize=(10, 6), color='white', edgecolor='black')
gdf.plot(ax=ax, color='blue',markersize=gdf['mag']*4, alpha=0.5)
```

```python
# Create a world map using the Natural Earth dataset
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
ax = world.plot(figsize=(10, 6), color='white', edgecolor='black')

# Add the earthquake data to the map
gdf.plot(ax=ax, color='red', markersize=gdf['mag']*2, alpha=0.5)

# Add a legend
plt.title('Earthquakes (Magnitude 4.0+)')
plt.legend(['Magnitude'])

# Display the map
plt.show()
```
